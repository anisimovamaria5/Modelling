```python
class Solution:
    def missing_number(self, nums: list[int]) -> int:
        nums_set = set(nums)
        n = len(nums)
        for x in range(n+1):
            if x not in nums_set:
                print(x)
                return x

nums = [3,1,0]
a = Solution()
print(a.missing_number(nums))
```

```python
from itertools import product
class Solution:
    def two_sum(self, nums: list[int], target:int) -> list[int]:
        n = len(nums)
        for i, j in product(range(n), repeat=2):
            if i != j:
                if nums[i] + nums[j] == target:
                    return [i,j]
        return []

nums = [3,2,4]
target = 6
a = Solution()
print(a.two_sum(nums, target))   
```

```python
from typing import Counter

class Solution:
    def long(self, s:str) -> int:
        count = Counter(s)
        length = 0
        has_odd = False
        for x in count.values():
            if x % 2 == 0:
                length += x
            else:
                length += x - 1
                has_odd = True
        if has_odd:
            length += 1
        return length



s = 'ccc'    
a = Solution()
print(a.long(s))
```

```python
class Solution:
    def intersection(self, nums1: list[int], nums2: list[int]) -> list[int]:
        l = len(nums1)
        lst = []
        for i in range(l):
            if nums1[i] in nums2:
                lst.append(nums1[i])
        return list(set(lst))
    

nums1 = [1,2,2,1]  
nums2 = [2,2]  
a = Solution()
print(a.intersection(nums1, nums2))
```

```python
class Solution:
    def insert_search(self, nums: list[int], target: int) -> int:
        l = len(nums)
        left = 0
        right = l - 1
        while left <= right:
            mid = (left + right) // 2
            if nums[mid] == target:
                return mid
            elif nums[mid] < target:
                left = mid + 1
            else:
                right = mid -1
        return left
    
nums =  [1,3,5,6] 
target = 4 
a = Solution()
print(a.insert_search(nums, target))
```

```python
from typing import Counter
class Solution:
    def max_power(self, s: str) -> int:
        max_power = 1
        current_power = 1
        for i in range(1, len(s)):
            if s[i] == s[i - 1]:
                current_power += 1
                max_power = max(max_power, current_power)
            else:
                current_power = 1 
        return max_power
    
    
s = "meet"
a = Solution()
print(a.max_power(s))
```

```python
import string
class Solution:
    def is_palindrome(self, s: str) -> bool:
        s = s.lower().replace(" ", '').translate(str.maketrans('','',string.punctuation))
        l = len(s)
        left = 0
        right = l - 1
        while left < right:
            if s[left] != s[right]:
                return False
            left += 1
            right -= 1
        return True

s = "race a car"
a = Solution()
print(a.is_palindrome(s))      
```

```python
class Solution:
    def is_valid_password(self, password: str) -> bool:
        cnt = 0
        if all(ord(c) < 128 for c in password):
            cnt += 1
        if any(char.isdigit() for char in password):
            cnt += 1
        if any(char.isupper() for char in password):
            cnt += 1
        if any(char in set('_#%') for char in password):
            cnt += 1
        if len(password) >= 8:
            cnt += 1
        if password.lower() != password.lower()[::-1]:
            cnt += 1
        if cnt == 6:
            return True
        else:
            return False

password = "zZzZ1234!"
a = Solution()
print(a.is_valid_password(password))             
```

```python
class Solution:
    def fib(self, n: int) -> int:
        if n <= 1:
            return n
        a, b = 0, 1
        for _ in range(2, n+1):
            a, b = b, a + b
        return b
        
        
n = 3
a = Solution()
print(a.fib(n))                 
```
